```python
from scipy.stats import genextreme as gev


t = np.linspace(-3000,3000,5000)
fit = gev.fit(data1)
print(fit)

fig, ax = plt.subplots(2, figsize=(8,6))
ax[0].plot(t, gev.pdf(t, *fit))
ax[1].plot(t, gev.cdf(t, *fit))

print(gev.cdf(np.max(data1)*2, *fit))
```
